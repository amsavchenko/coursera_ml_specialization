```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_digits, load_breast_cancer
```

```python
digits = load_digits()
```

```python
X_d = digits['data']
```

```python
y_d = digits['target']
```

```python
breast_cancer = load_breast_cancer()
```

```python
X_c = breast_cancer['data']
y_c = breast_cancer['target']
```

```python
from sklearn.cross_validation import cross_val_score
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
bernoulli = BernoulliNB()
multinomial = MultinomialNB()
gaussian = GaussianNB()
```

digits

```python
cross_val_score(bernoulli, X_d, y_d, cv=5, n_jobs=-1).mean()
```

```python
cross_val_score(multinomial, X_d, y_d, cv=5, n_jobs=-1).mean()
```

```python
cross_val_score(gaussian, X_d, y_d, cv=5, n_jobs=-1).mean()
```

breast cancer

```python
cross_val_score(bernoulli, X_c, y_c, cv=5, n_jobs=-1).mean()
```

```python
cross_val_score(multinomial, X_c, y_c, cv=5, n_jobs=-1).mean()
```

```python
cross_val_score(gaussian, X_c, y_c, cv=3, n_jobs=-1).mean()
```
